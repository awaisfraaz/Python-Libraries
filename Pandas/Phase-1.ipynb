```python
# Example: Assign grade based on marks without lambda function

def get_grade(marks):
    if marks <50:
        return "F"
    elif marks  <=60:
        return  "D"
    elif marks <=70:
        return "C"
    elif marks <=80:
        return "B"
    elif marks <=90:
        return "A"
    else:
        return "0"

df2["Grade"] = df2["mark"].apply(get_grade)
```
